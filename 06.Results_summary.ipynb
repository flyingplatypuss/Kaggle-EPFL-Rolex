```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# URLs for the prediction and validation data
url_pred_LR = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/LR_y_pred%20(1).csv'
url_val_LR = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/LR_y_val%20(1).csv'
url_pred_DT = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/DT_y_pred.csv'
url_val_DT = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/DT_y_val.csv'
url_pred_KNN = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/KNN_y_pred.csv'
url_val_KNN = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/KNN_y_val.csv'
url_pred_RF = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/RF_y_pred.csv'
url_val_RF = 'https://raw.githubusercontent.com/flyingplatypuss/Kaggle-EPFL-Rolex/main/RF_y_val.csv'

# Load the data
y_val_LR = pd.read_csv(url_val_LR)
y_val_KNN = pd.read_csv(url_val_KNN)
y_val_DT = pd.read_csv(url_val_DT)
y_val_RF = pd.read_csv(url_val_RF)
y_pred_LR = pd.read_csv(url_pred_LR)
y_pred_KNN = pd.read_csv(url_pred_KNN)
y_pred_DT = pd.read_csv(url_pred_DT)
y_pred_RF = pd.read_csv(url_pred_RF)

# Define a function to calculate all metrics
def calculate_metrics(y_true, y_pred):
    return {
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred, average='weighted'),
        'Recall': recall_score(y_true, y_pred, average='weighted'),
        'F1 Score': f1_score(y_true, y_pred, average='weighted')
    }

# Define a function to round the values of a dictionary to a specified precision
def round_dict_values(d, precision):
    return {k: round(v, precision) for k, v in d.items()}

# Calculate and round metrics to 4 decimal places
metrics_LR = round_dict_values(calculate_metrics(y_val_LR, y_pred_LR), 4)
metrics_KNN = round_dict_values(calculate_metrics(y_val_KNN, y_pred_KNN), 4)
metrics_DT = round_dict_values(calculate_metrics(y_val_DT, y_pred_DT), 4)
metrics_RF = round_dict_values(calculate_metrics(y_val_RF, y_pred_RF), 4)

# Create a DataFrame with the rounded metrics
df_metrics = pd.DataFrame({
    'Accuracy': [metrics_LR['Accuracy'], metrics_KNN['Accuracy'], metrics_DT['Accuracy'], metrics_RF['Accuracy'], round(0.58125, 4)],
    'Precision': [metrics_LR['Precision'], metrics_KNN['Precision'], metrics_DT['Precision'], metrics_RF['Precision'], round(0.6020, 4)],
    'Recall': [metrics_LR['Recall'], metrics_KNN['Recall'], metrics_DT['Recall'], metrics_RF['Recall'], round(0.58125, 4)],
    'F1 Score': [metrics_LR['F1 Score'], metrics_KNN['F1 Score'], metrics_DT['F1 Score'], metrics_RF['F1 Score'], round(0.5828, 4)]
}, index=['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'CamemBERT'])

# Transpose the DataFrame
df_metrics_trans = df_metrics.T


df_metrics_trans
```
